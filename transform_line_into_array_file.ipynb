```python

import pandas as pd   # here import the pandas library


dates = (                                                     #take the list of dates through JSON
    data["dimension"]["TLIST(D1)"]["category"]["index"]
)
category_ids =(                                                  #take the list of Category ID's through JSON
    data["dimension"]["C04132V04898"]["category"]["index"]
)
category_labels = (
    data["dimension"]["C04132V04898"]["category"]["label"]
)
values = data["value"]

# Build table
rows = []           #here create thge empty list
i = 0                              #start index pointer for reading the values one by one

for date in dates:
    for cat in category_ids:
        rows.append({                                #create one row for date, category, demand value,then add in the rows.
            "date": date,
            "category_id": cat,
            "category": category_labels[cat],
            "demand": values[i]
        })
        i += 1

df = pd.DataFrame(rows)       # convert the list of rows dict in the pandas dataframe
df.head()              #display first 5 rows.
```

```python
df = df.drop(columns=["category_id"])  # here drop the category id column
df.head()
```

```python
# here convert date into proper manner

def convert_date(x):
    yyyy = x[0:4]
    mm = x[5:7]
    dd = x[8:10]
    return f"{yyyy}-{mm}-{dd}"

df["date"] = (
    df["date"].apply(convert_date)
)
df.head()
```

```python
# now here saved the cleaned Json ARRAY file.
df.to_json(
     r"D:\mongo_data\gas_clean.json",
    orient="records",
    indent=4 
)
```
